```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, beta, expon
```

Let's start by creating a function that generates one sample of the game. Feel free to change this code as you find it suitable, for example by changing how the sampling is created or if you want to sample cards with replacement (this could be useful for the last questions).

```python
def sample_cards(N, n, k):
    """
    Generates a simulated sample of how the cards are shuffled among players. 
    
    Arguments: 
        - N : Total number of cards
        - n : Number of players
        - k : Number of cards per player
        
    Returns: 
        - 2D numpy array where each row represent each player and each column their card in ascending order. 
    """
    
    assert n * k <= N, "There are not enough cards for all the players!"
    
    _cards = np.random.choice(a=np.arange(1, N+1), size=n*k, replace=False)
    _cards = _cards.reshape(n,k)
    _cards.sort(axis=1)
    return _cards
```

### 2. What is the distribution of the largest gap?

For each simulation, found the largest distance between consecutive numbers in the game. To make this easy, let us first consider the largest gap among the pool of **all** the cards among players (if the largest gap is happening in the hand of one single player, that is ok). This will make the analysis more simple. 

```python
# To Do cell

def largest_gap(arr):
    """
    Finds largest gap between consecutive integers in a given numpy array 
    
    Arguments: 
        - arr : sorted 1D numpy array containing integers
        
    Returns: 
        - an integer representing the largest gap 
    """
    
    final_gap = 0
    
    # step through sorted array
    # keep track of largest gap
    for i in np.arange(0, len(arr) - 1):
        new_gap = arr[i+1] - arr[i]
        if new_gap > final_gap:
            final_gap = new_gap
    
    return final_gap
```

```python
def simulate_gaps_pooled(n_sims, N, n, k):

    # store results
    max_gaps_pooled = np.zeros(n_sims)

    # run simulations
    for i in np.arange(0, n_sims):
        # generate pooled card sample
        cards = sample_cards(N, n, k)
        cards_flat = cards.flatten()
        cards_flat_sorted = np.sort(cards_flat)

        # save largest gap
        max_gaps_pooled[i] = largest_gap(cards_flat_sorted)

    return max_gaps_pooled
    
plt.hist(simulate_gaps_pooled(10000, 100, 6, 3), bins = np.arange(1, N+1))
plt.xlabel("Max Gap (All Players)")
plt.ylabel("Count")
plt.title("Distribution of Max Gap Between All Players (N=100, n=6, k=3)")
```

Repeat the experiment but know compute the largest gap between cards of different players. This is the gap that actually matters for the purposes of the game: what is the maximum number of numbers that a player will have to wait before playing the next card?

```python
# To Do cell

def get_card_ids(cards_arr):
    """
    Sorts all cards and creates array with associated players
    
    Arguments:
        - cards_arr: 2D numpy array of cards, each row is a player
    
    Returns:
        - 1D array representing the sorted cards
        - 1D array containing the player ids for each card
    """
    
    # flatten cards into a 1D array
    cards_flat = cards_arr.flatten()
    
    # get initial list of player ids
    n_players = len(cards_arr)
    hand_size = len(cards_arr[0,:])
    player_ids = np.repeat(np.arange(0, n_players), hand_size)

    # sort cards and players based on the cards array
    ind = np.argsort(cards_flat)
    cards_flat_sorted = cards_flat[ind]
    player_ids_sorted = player_ids[ind]
    
    return cards_flat_sorted, player_ids_sorted
            

def largest_gap_diff_players(cards, ids):
    """
    Finds largest gap between consecutive integers across players in a given numpy array 
    
    Arguments: 
        - cards: sorted 1D numpy array containing integers
        - ids: player ids for each card in (cards) array
        
    Returns: 
        - an integer representing the largest gap between different players
    """
    
    final_gap = 0
    
    for i in np.arange(0, len(cards) - 1):
        new_gap = cards[i+1] - cards[i]
        # save new gap size if it is larger
        # also must be between different players
        if new_gap > final_gap and ids[i] != ids[i+1]:
            final_gap = new_gap
    
    return final_gap
```

```python
def simulate_gaps_players(n_sims, N, n, k):

    # store results
    max_gaps_players = np.zeros(n_sims)

    # run simulations
    for i in np.arange(0, n_sims):
        # generate player cards
        cards = sample_cards(N, n, k)

        # store largest gap
        cards_sorted, player_ids = get_card_ids(cards)
        max_gaps_players[i] = largest_gap_diff_players(cards_sorted, player_ids)

    return max_gaps_players

plt.hist(simulate_gaps_players(10000, 100, 6, 3), bins = np.arange(1, N+1))
plt.xlabel("Max Gap (Diff Players)")
plt.ylabel("Count")
plt.title("Distribution of Max Gap Between Diff Players (N=100, n=6, k=3)")
```

How does these compare? Run simulations for different choices of the number of cards and players and see when the difference between these two is larger. 

Compute the mean of the two distributions. When this difference is larger? How does this depends of the total number of players `n`?

```python
# To Do cell

# function to get both types of gaps
def get_gaps(cards):

    # preprocessing
    cards_sorted, player_ids = get_card_ids(cards)
    
    # get gaps
    pooled_gap = largest_gap(cards_sorted)
    player_gap = largest_gap_diff_players(cards_sorted, player_ids)
    
    return pooled_gap, player_gap
```

```python
# compare means
# different card numbers

# mean diffs
mean_diffs = np.zeros(3)

n_sims = 10000
n = 6
k = 3

N = 100
mean_diffs[0] = np.mean(simulate_gaps_pooled(n_sims, N, n, k)) - np.mean(simulate_gaps_players(n_sims, N, n, k))
N = 1000
mean_diffs[1] = np.mean(simulate_gaps_pooled(n_sims, N, n, k)) - np.mean(simulate_gaps_players(n_sims, N, n, k))
N = 10000
mean_diffs[2] = np.mean(simulate_gaps_pooled(n_sims, N, n, k)) - np.mean(simulate_gaps_players(n_sims, N, n, k))

mean_diffs
```

It appears that the mean gap size between all players tends to be larger than the mean gap size between different players. As the number of cards increases, the difference between these two gaps also increases.

```python
# run simulations for varying numbers of cards (N)
# for each number N cards, run 10000 simulations
# compute mean gap difference

n = 6
k = 3

Ns = np.arange(100, 1100, 100)
mean_diffs = np.zeros(len(Ns))

for i in range(0, len(Ns)):
    total_diff = 0
    
    for j in range(0, 10000):
        # compute differences in gaps
        cards = sample_cards(Ns[i], n, k)
        pooled_gap, player_gap = get_gaps(cards)
        total_diff += (pooled_gap - player_gap)
    
    mean_diffs[i] = total_diff / 10000

plt.scatter(x = Ns, y = mean_diffs)
plt.hlines(y = 0, xmin = 100, xmax = 1000, linestyles = "dashed", colors = "black")
plt.xlabel("Number of Cards (N)")
plt.ylabel("Mean (All Player Max Gap) - (Diff Player Max Gap)")
plt.title("Effect of Number of Cards on Gap Difference")
```

Thus, on average, the max gap between all players is larger than the max gap between different players. As the number of cards increases, the average difference between these two types of max gaps also increases.

### 3.1. Binomial Distribution 

Can you think about any way of deriving the binomial distribution? Based on the output of one single game (function `sample_game()`), try to derive something that has Binomial distribution for certain parameter $p$. Run multiple simulations to see this and compare the obtained empirical distribution with the distribution of the binomial distribution (check `numpy.random.binomial` and `scipy.stats.binom`). 

We can simulate binomial distributions by checking if cards in a certain range appear in the players' hands. Suppose we have 100 cards total and we sample player cards with replacement. Suppose we care about whether players draw cards that are less than or equal to 20. Each draw is a Bernoulli trial with $p=0.2$. We can sum the successes from these Bernoulli trials to get a binomial draw. We demonstrate this method below, using a slightly modified `sample_cards` function that draws with replacement:

```python
# To Do cell
def sample_cards_replace(N, n, k):
    """
    Generates a simulated sample of how the cards are shuffled among players. (with replacement)
    
    Arguments: 
        - N : Total number of cards
        - n : Number of players
        - k : Number of cards per player
        
    Returns: 
        - 2D numpy array where each row represent each player and each column their card in ascending order. 
    """
        
    _cards = np.random.choice(a=np.arange(1, N+1), size=n*k, replace=True)
    _cards = _cards.reshape(n,k)
    _cards.sort(axis=1)
    return _cards
```

```python
# simulate a binomial distribution with n = 100, p = 0.2

n_sims = 10000
binom_draws = np.zeros(n_sims)

for i in range(0, n_sims):
    cards = sample_cards_replace(100, 25, 4) # 25 players * 4 cards/player = 100 cards
    cards = cards <= 20
    binom_draws[i] = np.sum(cards)

plt.hist(binom_draws, alpha = 0.5, bins = np.arange(0, 101), label = "The Mind", density = True)
plt.bar(np.arange(0, 101), binom.pmf(np.arange(0, 101), 100, 0.2), label = "scipy", alpha = 0.5, color = "orange")
plt.xlabel("Number of Sucesses")
plt.ylabel("Frequency")
plt.title("Simulated Binomial Distribution")
plt.legend()
```

### 3.2. Beta Distribution

The [order statistics](https://en.wikipedia.org/wiki/Order_statistic) of the uniform distribution are such that follow a Beta distribution. If well here we are not exactly sampling real numbers in $[0,1]$ following the uniform distribution, you can imagine that sampling one single number without repetition from $\{ 1, 2, 3, \ldots, N \}$ is not very different...

Compare the distribution of the lowest number in the sample (Part 1) with a [Beta distribution](https://en.wikipedia.org/wiki/Beta_distribution) with adequate choices of parameters. In order to do this, notice that the $k$th order statistics of a sample of $n$ uniformly distributed points in $[0,1]$ is $\text{Beta}(k, n+1-k)$ (These $n$ and $k$ are not the same as before, you need to pick the right values of these).

```python
# To Do cell
# 1st order statistic of N uniformly distributed points in [0,1] is Beta(1, N)

# Sampling without replacement

N = 100 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
X1s = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards(N, n, k)
    X1s[i] = np.min(cards)

no_replace_mean = np.mean(X1s)
    
# get Beta pdf
xs = np.linspace(0, 1, 1000)
betas = beta.pdf(xs, 1, n*k)

plt.figure()
plt.hist(X1s, bins = np.arange(1, N+1), density = True, alpha = 0.5, label = "The Mind")
plt.plot(xs*N, betas/N, label = "scipy")
plt.xlabel("Lowest Number")
plt.ylabel("Density")
plt.title("Simulated Beta Distribution (Without Replacement)")
plt.legend()

# Sampling with replacement

N = 100 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
X1s = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards_replace(N, n, k)
    X1s[i] = np.min(cards)

replace_mean = np.mean(X1s)
    
# get Beta pdf
xs = np.linspace(0, 1, 1000)
betas = beta.pdf(xs, 1, n*k)

plt.figure()
plt.hist(X1s, bins = np.arange(1, N+1), density = True, alpha = 0.5, label = "The Mind")
plt.plot(xs*N, betas/N, label = "scipy")
plt.xlabel("Lowest Number")
plt.ylabel("Density")
plt.title("Simulated Beta Distribution (With Replacement)")
plt.legend()
```

### 3.3. Exponential Distribution

If we call $U_1$ the minimum number in the sample (Part 1), then make an histogram with the distribution of the transformed random variable given by 
$$
X = - \ln \left( 1 - \frac{U_1}{N} \right)
$$
and compare this with the distribution of an exponential random variable with parameter $nk$, the total number of cards in the game. 

```python
# Sampling without replacement

N = 100 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
Xs = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards(N, n, k)
    U1 = np.min(cards)
    Xs[i] = np.log(1 - (U1 / N)) * -1

# get Exponential pdf
xs = np.linspace(0, 1, 2000)
exps = expon.pdf(xs, scale = 1/(n*k))

plt.figure()
plt.hist(Xs, density = True, bins = np.linspace(-np.log(1-(1/N)), -np.log(1-((N-1)/N)), N), alpha = 0.5, label = "The Mind")
plt.plot(xs, exps, label = "scipy")
plt.xlim(left = 0, right = 0.5)
plt.xlabel("Transformed Minimum")
plt.ylabel("Density")
plt.title("Simulated Exponential Distribution (Without Replacement)")
plt.legend()


# Sampling with replacement

N = 100 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
Xs = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards_replace(N, n, k)
    U1 = np.min(cards)
    Xs[i] = np.log(1 - (U1 / N)) * -1

# get Exponential pdf
xs = np.linspace(0, 1, 2000)
exps = expon.pdf(xs, scale = 1/(n*k))

plt.figure()
plt.hist(Xs, density = True, bins = np.linspace(-np.log(1-(1/N)), -np.log(1-((N-1)/N)), N), alpha = 0.5, label = "The Mind")
plt.plot(xs, exps, label = "scipy")
plt.xlim(left = 0, right = 0.5)
plt.xlabel("Transformed Minimum")
plt.ylabel("Density")
plt.title("Simulated Exponential Distribution (With Replacement)")
plt.legend()
```

### 4. What happens when $n * k << N$?

Some of the distribution we saw in previous sections are just approximate. Study what happens when we increase the number of cards. Try to understand how the cases of sampling with and without reposition compare as $n * k << N$. What do you expect to see in these cases? When the difference between the two sampling strategies will lead to similar or different results. 

```python
# To Do cell
# Simulating Beta with n*k << N
# Without replacement

N = 10000 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
X1s = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards(N, n, k)
    X1s[i] = np.min(cards)

# get Beta pdf
xs = np.linspace(0, 1, 1000)
betas = beta.pdf(xs, 1, n*k)

plt.figure()
plt.hist(X1s, bins = np.arange(1, N+1), density = True, alpha = 0.5, label = "The Mind")
plt.plot(xs*N, betas/N, label = "scipy")
plt.xlabel("Lowest Number")
plt.ylabel("Density")
plt.title("Simulated Beta Distribution, nk<<N (Without Replacement)")
plt.legend()

# Simulating Beta with n*k << N
# With replacement

N = 10000 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
X1s = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards_replace(N, n, k)
    X1s[i] = np.min(cards)

# get Beta pdf
xs = np.linspace(0, 1, 1000)
betas = beta.pdf(xs, 1, n*k)

plt.figure()
plt.hist(X1s, bins = np.arange(1, N+1), density = True, alpha = 0.5, label = "The Mind")
plt.plot(xs*N, betas/N, label = "scipy")
plt.xlabel("Lowest Number")
plt.ylabel("Density")
plt.title("Simulated Beta Distribution nk<<N (With Replacement)")
plt.legend()
```

```python
# Simulated exponential
# With n*k << N
# Without replacement

N = 10000 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
Xs = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards(N, n, k)
    U1 = np.min(cards)
    Xs[i] = np.log(1 - (U1 / N)) * -1

# get Exponential pdf
xs = np.linspace(0, 1, 2000)
exps = expon.pdf(xs, scale = 1/(n*k))

plt.figure()
plt.hist(Xs, density = True, bins = np.linspace(-np.log(1-(1/N)), -np.log(1-((N-1)/N)), N), alpha = 0.5, label = "The Mind")
plt.plot(xs, exps, label = "scipy")
plt.xlim(left = 0, right = 0.5)
plt.xlabel("Transformed Minimum")
plt.ylabel("Density")
plt.title("Simulated Exponential Distribution, nk<<N (Without Replacement)")
plt.legend()

# Simulated exponential
# With n*k << N
# With replacement

N = 10000 # number of cards
n = 6 # number of players
k = 3 # cards per player

# get The Mind distribution
n_sims = 100000
Xs = np.zeros(n_sims)
for i in range(0, n_sims):
    cards = sample_cards_replace(N, n, k)
    U1 = np.min(cards)
    Xs[i] = np.log(1 - (U1 / N)) * -1

# get Exponential pdf
xs = np.linspace(0, 1, 2000)
exps = expon.pdf(xs, scale = 1/(n*k))

plt.figure()
plt.hist(Xs, density = True, bins = np.linspace(-np.log(1-(1/N)), -np.log(1-((N-1)/N)), N), alpha = 0.5, label = "The Mind")
plt.plot(xs, exps, label = "scipy")
plt.xlim(left = 0, right = 0.5)
plt.xlabel("Transformed Minimum")
plt.ylabel("Density")
plt.title("Simulated Exponential Distribution, nk<<N (With Replacement)")
plt.legend()
```

We note that when $nk<<N$, the distribution approximations get much better. It is hard to tell whether there are any differences between the two sampling strategies, so we compute means for the two sampling strategies for varying values of $nk$ (we do this for exponential only):

```python
# Simulated exponential
# With and without replacement

n = 6 # number of players
k = 3 # cards per player
Ns = np.arange(20, 1000, 10)
n_sims = 1000

mean_no_replace = np.zeros(len(Ns))
mean_replace = np.zeros(len(Ns))

for i in range(0, len(Ns)):
    sum_no_replace = 0
    sum_replace = 0
    
    for j in range(0, n_sims):
        cards_no_replace = sample_cards(Ns[i], n, k)
        cards_replace = sample_cards_replace(Ns[i], n, k)
        
        U1_no_replace = np.min(cards_no_replace)
        U1_replace = np.min(cards_replace)
        
        X_no_replace = np.log(1 - (U1_no_replace / Ns[i])) * -1
        X_replace = np.log(1 - (U1_replace / Ns[i])) * -1
        
        sum_no_replace += X_no_replace
        sum_replace += X_replace
    
    mean_no_replace[i] = sum_no_replace / n_sims
    mean_replace[i] = sum_replace / n_sims

plt.scatter(Ns, mean_no_replace, label = "Without Replacement")
plt.scatter(Ns, mean_replace, label = "With Replacement")
plt.plot(Ns, mean_no_replace, label = "Without Replacement")
plt.plot(Ns, mean_replace, label = "With Replacement")
plt.xlabel("Number of Cards (N)")
plt.ylabel("Mean of Simulated Exponential")
plt.title("Effect of Number of Cards on Replacement Strategies")
plt.legend()
```

It appears that for smaller $N$, the with replacement strategy tends to draw larger transformed values. As $N$ gets larger, the difference between the two strategies appears to decrease.
